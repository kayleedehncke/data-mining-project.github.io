```python
import pandas as pd
import numpy as np
import dataframe_image as dfi
import os

from collections import Counter
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import fpgrowth, association_rules
from graphviz import Digraph
```

```python
df = pd.read_csv("../data/final_dataset.csv")

print("Shape:", df.shape)
df.head()
```

```python
print(df.columns.tolist())
print("\nData types:")
print(df.dtypes)
```

```python
selected_numeric = [
    "TMAX",
    "p_pov",
    "p_treec",
    "p_asthma",
    "overall_score",
    "hhb_score"
]

target_col = "heat_related_any"

df_fp = df[["date"] + selected_numeric + [target_col]].copy()

print("Working shape:", df_fp.shape)
df_fp.head()
```

```python
df_fp["date"] = pd.to_datetime(df_fp["date"], errors="coerce")
df_fp["month"] = df_fp["date"].dt.month

df_fp = df_fp[df_fp["month"].isin([5, 6, 7, 8, 9])].copy()

print("After warm-month filter:", df_fp.shape)
df_fp.head()
```

```python
print("Before preprocessing snapshot:")
dfi.export(
    df_fp.head(10), 
    "../../website/resources/img/tables/fp_raw_table.png",
    table_conversion="chrome",
    max_cols = 10
)
display(df_fp.head())
```

```python
def quantile_item(series, prefix):
    binned = pd.qcut(series, q=2, labels=["low", "high"], duplicates="drop")
    return prefix + "_" + binned.astype(str)
```

```python
item_df = pd.DataFrame(index=df_fp.index)

for col in selected_numeric:
    item_df[col] = quantile_item(df_fp[col], col)

item_df.head()
```

```python
print("After preprocessing snapshot:")
dfi.export(
    item_df.head(10),
    "../../website/resources/img/tables/fp_clean_table.png",
    table_conversion="chrome",
    max_cols = 10
)
display(item_df.head())
```

```python
transactions = []

for idx, row in item_df.iterrows():
    tx = [
        row["TMAX"],
        row["p_pov"],
        row["p_treec"],
        row["p_asthma"],
        row["overall_score"],
        row["hhb_score"]
    ]
    
    if df_fp.loc[idx, target_col] == 1:
        tx.append("heat_event_yes")
    
    transactions.append(tx)

print("Number of transactions:", len(transactions))
print("First 5 transactions:")
for tx in transactions[:5]:
    print(tx)
```

```python
item_counts = Counter(item for tx in transactions for item in tx)
item_counts.most_common(20)
```

```python
te = TransactionEncoder()
te_array = te.fit(transactions).transform(transactions)
basket = pd.DataFrame(te_array, columns=te.columns_)

print("Basket shape:", basket.shape)
basket.head()
```

```python
frequent_itemsets = fpgrowth(
    basket,
    min_support=0.001,
    use_colnames=True,
    max_len=2
)

frequent_itemsets["itemset_length"] = frequent_itemsets["itemsets"].apply(len)
frequent_itemsets = frequent_itemsets.sort_values(
    by=["support", "itemset_length"],
    ascending=[False, True]
)

print("Number of frequent itemsets:", len(frequent_itemsets))
frequent_itemsets.head(20)
```

```python
frequent_itemsets_display = frequent_itemsets.copy()
frequent_itemsets_display["itemsets"] = frequent_itemsets_display["itemsets"].apply(
    lambda x: ", ".join(sorted(list(x)))
)

frequent_itemsets_display.head(20)
```

```python
dot = Digraph(comment="FP-Growth Itemset Graph")

# 1. Add nodes for 1-item itemsets
for _, row in frequent_itemsets.iterrows():
    items = list(row["itemsets"])
    
    if len(items) == 1:
        item = items[0]
        label = f"{item}\\nSupport: {row['support']:.3f}"
        dot.node(item, label)

# 2. Add edges for 2-item itemsets
for _, row in frequent_itemsets.iterrows():
    items = list(row["itemsets"])
    
    if len(items) == 2:
        a, b = items
        edge_label = f"{row['support']:.3f}"
        dot.edge(a, b, label=edge_label)

# Render
dot.render("../../website/resources/img/plots/fp_growth_graph", format="png", cleanup=True)
```

```python
rules = association_rules(
    frequent_itemsets,
    metric="confidence",
    min_threshold=0.01
)

print("Number of rules:", len(rules))
rules.head()
```

```python
heat_rules = rules[
    rules["consequents"].apply(lambda x: "heat_event_yes" in x)
].copy()

heat_rules["antecedents"] = heat_rules["antecedents"].apply(
    lambda x: ", ".join(sorted(list(x)))
)
heat_rules["consequents"] = heat_rules["consequents"].apply(
    lambda x: ", ".join(sorted(list(x)))
)

heat_rules = heat_rules.sort_values(
    by=["lift", "confidence", "support"],
    ascending=[False, False, False]
)

heat_rules[["antecedents", "consequents", "support", "confidence", "lift"]].head(20)
```

```python
heat_itemsets = frequent_itemsets[
    frequent_itemsets["itemsets"].apply(lambda x: "heat_event_yes" in x)
].copy()

dot = Digraph(comment="FP-Growth Heat Event Itemset Graph")

# 1. Add nodes for 1-item itemsets
for _, row in heat_itemsets.iterrows():
    items = list(row["itemsets"])
    
    if len(items) == 1:
        item = items[0]
        label = f"{item}\\nSupport: {row['support']:.3f}"
        dot.node(item, label)

# 2. Add edges for 2-item itemsets
for _, row in heat_itemsets.iterrows():
    items = list(row["itemsets"])
    
    if len(items) == 2:
        a, b = items
        edge_label = f"{row['support']:.3f}"
        dot.edge(a, b, label=edge_label)


# Render the tree
dot.render("../../website/resources/img/plots/fp_growth_tree_heat_event", format="png", cleanup=True)
```

```python
heat_itemsets["itemsets"] = heat_itemsets["itemsets"].apply(lambda x: ", ".join(sorted(list(x))))
heat_itemsets.sort_values("support", ascending=False).head(20)
```

```python
heat_rules = rules[
    rules["consequents"].apply(lambda x: "heat_event_yes" in x)
].copy()

print("Number of rules with heat_event_yes as consequent:", len(heat_rules))

if len(heat_rules) > 0:
    heat_rules["antecedents"] = heat_rules["antecedents"].apply(
        lambda x: ", ".join(sorted(list(x)))
    )
    heat_rules["consequents"] = heat_rules["consequents"].apply(
        lambda x: ", ".join(sorted(list(x)))
    )

    heat_rules = heat_rules.sort_values(
        by=["lift", "confidence", "support"],
        ascending=[False, False, False]
    )

display(
        heat_rules[[
            "antecedents",
            "consequents",
            "antecedent support",
            "consequent support",
            "support",
            "confidence",
            "lift"
        ]].head(20)
)

dfi.export(
    heat_rules[[
        "antecedents",
        "consequents",
        "antecedent support",
        "consequent support",
        "support",
        "confidence",
        "lift"
    ]].head(20),
    "../../website/resources/img/tables/fp_heat_rules.png",
    table_conversion="chrome",
    max_cols = 10
)
```

```python
dot = Digraph(comment="FP-Growth Heat Event Rule Graph")

# Add nodes and edges based on the rules table
for _, row in heat_rules.iterrows():
    antecedents = row["antecedents"].split(", ")
    consequent = row["consequents"]
    support = row["support"]
    confidence = row["confidence"]
    lift = row["lift"]

    # Add nodes for antecedents and consequent
    for antecedent in antecedents:
        dot.node(antecedent, antecedent)
    dot.node(consequent, consequent)

    # Add edges from antecedents to consequent
    for antecedent in antecedents:
        edge_label = f"Support: {support:.3f}\\nConfidence: {confidence:.3f}\\nLift: {lift:.3f}"
        dot.edge(antecedent, consequent, label=edge_label)



# Render the tree
dot.render("../../website/resources/img/plots/fp_growth_tree_heat_rule", format="png", cleanup=True)
```

Because heat_event_yes was rare in the dataset, initial rule generation using a higher confidence threshold produced no rules with the event as the consequent. This was addressed by lowering the confidence threshold and examining lift, which is more appropriate for rare outcomes. The resulting itemsets showed that high vulnerability indicators such as high overall score, high HHB score, high poverty, and high temperature were the variables most associated with heat-related events.
